```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/diabetes.csv")
data
```

```python
for col in data.columns:
  print(f"column : {col}")
  print (f"unique values : {data[col].nunique()}")
  print(data[col].value_counts())
  print("-------------")
```

```python
for column in data.columns:
  plt.figure(figsize=(8,5))
  data.boxplot([column])
  plt.title(f"boxplot of {column}")
  plt.show()
```

```python
data.columns
```

```python
glucose_median=data['Glucose'].median()
data['Glucose']=data['Glucose'].replace(0,glucose_median)
```

```python
blood_median=data['BloodPressure'].median()
data['BloodPressure']=data['BloodPressure'].replace(0,blood_median)
```

```python
inulin_median=data['Insulin'].median()
data['Insulin']=data['Insulin'].replace(0,inulin_median)
```

```python
q1=data['Insulin'].quantile(0.25)
q3=data['Insulin'].quantile(0.75)
iqr=q3-q1
upper_bound=q3+(1.5*iqr)
data['Insulin']=np.where(data['Insulin']>upper_bound,upper_bound,data['Insulin'])
upper_bound
```

```python
plt.figure(figsize=(8,5))
data.boxplot(['Insulin'])
plt.title("boxplot of Insulin")
plt.show()
```

```python
matrix=data.corr()
plt.figure(figsize=(15,10))
sns.heatmap(matrix,annot=True,cmap="coolwarm")
plt.show()
```
